```python
# add a feature with intial value 1
df['Quantity']=1
```

```python
df.head()
```

```python
# grouping of two features by adding the quantity 
# unstack the values then set index of member_number
df.groupby(['Member_number','itemDescription'])['Quantity'].sum()
# .unstack().reset_index().set_index('Member_number')
```

```python
transcations= df.groupby(['Member_number', 'itemDescription'])['Quantity'].sum().unstack().reset_index().set_index('Member_number')
```

```python
transcations
```

```python
# fill NAN values with 0
transcations= transcations.fillna(0)
```

```python
transcations
```

```python
# function to set value 0 where df is greater than 0
def encode(df):
    return 1 if df>0 else 0
basket= transcations.applymap(encode)
```

```python
basket
```

```python
# import apriori and assocation rules (machine learning extension)
from mlxtend.frequent_patterns import apriori
```

```python
from mlxtend.frequent_patterns import association_rules
```

```python
frequent_itemset= apriori(basket, min_support=0.06, use_colnames=True)
# items which appear frequents
```

```python
frequent_itemset
```

```python
rules= association_rules(frequent_itemset, metric='lift', min_threshold=1)
```

```python
rules.head()
```

```python
# keep only those items where condition satisfied
rules[(rules['confidence']>0.4) & (rules['lift']>1)]
```
